```python
import pandas as pd
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('times_search.csv', names=['size', 'Search', 'time'])
df= data.pivot(index='size', columns='Search', values='time')
```

```python
import matplotlib
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

temp_plot = df.plot(figsize=(10, 8), logx=False, logy=True)
temp_plot.set_ylabel('Time, nanoseconds', fontsize=15)
temp_plot.set_xlabel('Size', fontsize=15)
temp_plot.set_title('Logarithmic plot of search timing', fontsize=15)
plt.savefig('timing_search_graph.png')
```
